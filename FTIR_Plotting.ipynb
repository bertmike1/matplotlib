```python
import pandas as pd
import matplotlib.pyplot as plt
cellulose = pd.read_csv('/content/untreated cellulose.CSV')
sci_1min = pd.read_csv('/content/Cellulose+Vanillin 6.9mg 1min.CSV')
sci_10min = pd.read_csv('/content/Cellulose+Vanillin 6.8mg 10min.CSV')
sci_20min = pd.read_csv('/content/Cellulose+Vanillin 6.9mg 20min.CSV')
sci_30min = pd.read_csv('/content/Cellulose+Vanillin 6.4mg 30min.CSV')
cellulose.columns = ['Wavenumber', 'Absorbance']
sci_1min.columns = ['Wavenumber', 'Absorbance']
sci_10min.columns = ['Wavenumber', 'Absorbance']
sci_20min.columns = ['Wavenumber', 'Absorbance']
sci_30min.columns = ['Wavenumber', 'Absorbance']
plt.plot(cellulose.Wavenumber, cellulose.Absorbance)
plt.plot(sci_1min.Wavenumber, sci_1min + .15)
plt.plot(sci_10min.Wavenumber, sci_10min + .30)
plt.plot(sci_20min.Wavenumber, sci_10min + .45)
plt.plot(sci_30min.Wavenumber, sci_10min + .60)
plt.xlabel('Wavenumber (cm-1)')
plt.ylabel('Absorbance')
plt.title('Cellulose FTIR')
plt.xlim(4000, 400)
plt.ylim(-0.04, 0.8)
plt.show()
```
